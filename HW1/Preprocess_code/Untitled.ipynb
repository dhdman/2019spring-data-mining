```python
#!/usr/bin/env python
# coding: utf-8
# In[1]:
from selenium import webdriver
import numpy as np
import pandas as pd
import time
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
from pathlib import Path, PureWindowsPath
from selenium.webdriver.chrome.options import Options
# In[3]:
```

```python
import os

# 將抓取的資料存取成路徑 再用 os.listdir 抓去路徑內各個檔案之路徑+名稱

download_path = "C:\\Users\\USER\\github\\HW1--data-mining\\HW1\\ETF_data_set\\filter_data"
file_path_list = []
file_name =[]
for file in os.listdir(download_path):
    file_path_list.append(os.path.join(download_path, file))
    file_name.append(file)
```

```python
def load_data (etfcsvpath) :
    etfdata = pd.read_csv(etfcsvpath)
    return etfdata
```

```python
count = 0
    # 將指定路徑ETF檔案run一個for loop 目的是抓取每個etf裡面的各個名稱
for path  in file_path_list  :
    etfdata = pd.read_csv(PureWindowsPath(path))
    
    # 指定下載檔案存放路徑
    download_path = os.getcwd() + file_name[count]
    count +=1
    # 檢查路徑是否存在
    if os.path.exists(download_path) and os.path.isdir(download_path):
        print(download_path + '資料夾已存在')
    else:
        os.makedirs(download_path)
        print('已新增資料夾')

    # 設定下載檔案存放路徑
    prefs = {"download.default_directory" : download_path}
    options = webdriver.ChromeOptions()
    options.add_experimental_option("prefs",prefs)

    # 將存放路徑清空
    for file in os.listdir(download_path):
        file_path = os.path.join(download_path, file)
        try:
            if os.path.isfile(file_path):
                os.unlink(file_path)
        except Exception as e:
            print(e)
            
            
    driver = webdriver.Chrome('C:\\Users\\USER\\Desktop\\chromedriver1',options=options)
    
    print(path)
    
    # Symbol 為各個ETF裡面的名稱columns
    
    # 以下是使用selenium select by css 抓取elements 在 run for loop catch all element on this css
    
    
    for symbol in etfdata['Symbol']:
        print(symbol)
        url = 'https://finance.yahoo.com/quote/{}/history?period1=1110384000&period2=1552665600&interval=1wk&filter=history&frequency=1wk'.format(symbol)
        driver.get(url)
    
    # sleep 是為了在打開網頁時 先讓網頁跑出來的等待時間 避免程式太快速執行指令導致抓取不到目標ERROR
    
        time.sleep(3)
        
    # select historical 
        try :
            arrows = driver.find_elements_by_css_selector (".historical div div span svg")
            arrows [0].click ()
            time.sleep (4)
        except :
            url = 'https://finance.yahoo.com/quote/{}/history?period1=1110384000&period2=1552665600&interval=1wk&filter=history&frequency=1wk'.format(symbol)
            driver.get(url)
            time.sleep(3)
            try :
                arrows = driver.find_elements_by_css_selector (".historical div div span svg")
                arrows [0].click ()
                time.sleep (4)
            except :
                print('error' + symbol)
                pass
        # select timetrend term
    
        durations = driver.find_elements_by_css_selector ("[data-test=\"date-picker-menu\"] div span")
        for duration in durations :
            #print('time_span = ', duration.text)
            if duration.text == "MAX" :
                duration.click ()
                time.sleep (4)
                break
            
            
            
        # select done buttom
    
        buttons = driver.find_elements_by_css_selector ("[data-test=\"date-picker-menu\"] div button")
        for button in buttons :
            #print('buttom = ',button.text)
            if button.text == "Done" :
                button.click ()
                time.sleep (4)
                break
            
            
        # select apply buttom
    
        buttons = driver.find_elements_by_css_selector ("button span")
        for button in buttons :
            #print('buttom = ',button.text)
            if button.text == "Apply" :
                button.click ()
                time.sleep (4)
                break
            
            
            
        # select download 
    
        links = driver.find_elements_by_css_selector ("a span")
        for link in links :
            #print('link =',link.text)
            if link.text == "Download Data" :
                #link.click ()                      #<===========  download already so we don't run it
                time.sleep (4)   
                break
                
    #driver.quit()
```
