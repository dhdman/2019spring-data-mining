```python
import numpy as np
import datetime as dt
import pandas as pd
```

```python
data = pd.read_csv('C:\\Users\\USER\\github\\HW1--data-mining\\HW1\\ETF_data_set\\Merge_all_data.csv')

data_drop = data.dropna(axis=0)

data_drop = data_drop.reset_index()

data_drop = data_drop.iloc[:,2:]
```

```python
data_drop
```

```python
def return_rate(etf_name,return_list) :
    for i in range(len(data_drop[etf_name])-1) :
        return_list.append((data_drop[etf_name][i+1] - data_drop[etf_name][i]) / data_drop[etf_name][i])
    return return_list
        
```

### filter into 2016-01-01 to 2018-12-01

```python
start = dt.datetime(2015,1,1)
end = dt.datetime(2019,1,1)

listtime = data_drop['Date']
listdate = []
for time in listtime :
     listdate.append(dt.datetime.strptime(time,'%Y-%m-%d'))

data_drop['Date'] = listdate    ### trun it into datetime


filter = (data_drop['Date'] <= end )
clean_data = data_drop[filter]
filter = (clean_data['Date'] >= start)
clean_data = clean_data[filter]
```
