```python
import sympy
import math
from scipy.optimize import fsolve
import time
import matplotlib.pyplot as plt
```

```python
# test data
#arr_returns = [0.01, 0.02, -0.01, 0.02, 0.04, 0.03, 0.015, 0.01, 0.02, 0.03]
arr_returns = [0.01, 0.02, -0.01]
```

```python
arr_x = []
arr_y = []
for x in range(100):
    if x == 0:
        continue
    x = x/100
    arr_x.append(x)
    s = 0
    for r in arr_returns:
        s = s + math.exp(-1*r/x)
    y = s / len(arr_returns)
    arr_y.append(y)
```

```python
%matplotlib inline
plt.plot(arr_x,arr_y)
plt.xlabel('Riskiness')
plt.ylabel('Expectation')
plt.show()
```
