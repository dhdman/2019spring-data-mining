```python
import sympy
import math
from scipy.optimize import fsolve
import time
import matplotlib.pyplot as plt
import pandas as pd
```

```python
import numpy  as  np
rlist = np.arange(0,100,10)
```

```python
def cal_riskiness_index(arr_returns):
    arr_x = []
    arr_y = []
    for x in rlist:  #<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<<< i change
        if x == 0:
            continue
        x = x
        arr_x.append(x)
        s = 0
        for r in arr_returns:
            s = s + math.exp(-1*r/x)
        y = s / len(arr_returns)
        arr_y.append(y)
    return(arr_x,arr_y)
```
