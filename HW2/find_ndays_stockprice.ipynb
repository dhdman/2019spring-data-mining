```python
import os
import jieba
import numpy as np
import pandas as pd
from collections import Counter
```

## 將新聞日期存取成datelist

```python
datelist = []
dir_path = './ctee_TSMC'
for file in os.listdir(dir_path):
    file_path = os.path.join(dir_path,file)
    datelist.append(file_path[-14:-4])
```

## 讀取datelist區間股價

```python
stock_price = pd.read_csv('./TSMC_price_daily/stock_price_2016120020190413.csv',encoding= "big5")
#stock_price = pd.read_csv('C:/Users/USER/github/HW1--data-mining/HW2/TSMC_price_daily/stock_price_2016120020190413.csv',encoding= "big5")
```

```python
stock_price.iloc[0:10,:]
```

## 將datelist 轉成datetime.dtype

```python
from datetime import datetime ,timedelta
datetimelist = []
for date in datelist :
    datetimelist.append(datetime.strptime(date,"%Y_%m_%d"))
datetimelist[:3]
```

## findbusiness day

```python
#pip install business_calendar
from business_calendar import Calendar, MO, TU, WE, TH, FR
cal = Calendar()
te = cal.addbusdays(datetimelist[0],5)
te
```

## 將時間軸對齊

```python
datetimelist01 = datetimelist[:-4]
```

## 將datelist 延後四天(business)

```python
datenow = []
for i in range(len(datetimelist)) :
    datenow.append(cal.addworkdays(datetimelist[i],-1))

datefive = []
for i in range(len(datenow)-5) :
    datefive.append(cal.addworkdays(datenow[i],5))

datefives = []
for i in range(len(datenow)-4) :
    datefives.append(datenow[i] + timedelta(7))
```

## Filter price by datelist(five days later)

```python
stock_date = []
for date in stock_list :
    stock_date.append(datetime.strptime(date,"%Y/%m/%d"))
stock_date[:3]
stock_price['年月日'] = stock_date
```
